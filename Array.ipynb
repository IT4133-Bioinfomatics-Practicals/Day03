```python
arr5 = np.array([1, 2, 3, 4, 5])
arr5

print("First element:", arr5[0])
print("Second element:", arr5[1])
```

```python
sub_arr1 = arr5[0:3]
print("Sub-array:", sub_arr1)

range_val = max(sub_arr1) - min(sub_arr1)
print("Range of elements from 1st to 3rd:", range_val)
```

```python
sub_arr2 = arr5[1:]
print("Sub array: ", sub_arr2)

range_val2 = max(sub_arr2) - min(sub_arr2)
print("Range of elements from 2nd element:", range_val2)
```

```python
sub_arr3 = arr5[-2:]
print("Sub array: ", sub_arr3)

range_val3 = max(sub_arr3) - min(sub_arr1)
print("Range of elements from index - 2:", range_val3)
```

```python
import numpy as np

arr1 = [1,2,3]
arr2 = [4,5,6]

arr_2D = np.array([arr1, arr2])
arr_2D
```

```python
print("1st row 2nd col element:", arr_2D[0,1])
print("2nd row elements:", arr_2D[1])
print("3rd column elements:", arr_2D[:,2])

range_row1 = max(arr_2D[0]) - min(arr_2D[0])
print("Range of elements in 1st row:", range_row1)
```

```python
sub_arr = arr_2D[0:2, 1:3]
print("Sub array:\n", sub_arr)

range_val = np.max(sub_arr) - np.min(sub_arr)
print("Range of element in 1st and 2nd row with 2nd and 3rd col:",range_val)
```
